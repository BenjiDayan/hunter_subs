```python
def split_text_lines(text):
    lines = text.split('\n')
    out = []
    for line in lines:
        if line[0] == '{':
            out.append(line)
        else:
            out[-1] += '\\n' + line

    return out

out = split_text_lines(out_simplified)
out
with open('ep1_out_26_08_2024_subset_gemini_simplified.sub', 'w') as file:
    for line in out:
        file.write(line + '\n')
```
